```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = tf.Variable(tf.linspace(-3.5, 3.5, 1000))

with tf.GradientTape() as tape:
  sigmoid = tf.nn.sigmoid(xs)
sigmoid_grad = tape.gradient(sigmoid, [xs])[0]

with tf.GradientTape() as tape:
  relu = tf.nn.relu(xs)
relu_grad = tape.gradient(relu, [xs])[0]

with tf.GradientTape() as tape:
  elu = tf.nn.elu(xs)
elu_grad = tape.gradient(elu, [xs])[0]

with tf.GradientTape() as tape:
  selu = tf.nn.selu(xs)
selu_grad = tape.gradient(selu, [xs])[0]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

axs[0].set_title("Function", fontsize=18)
axs[0].plot(xs.numpy(), sigmoid.numpy(), label="Sigmoid")
axs[0].plot(xs.numpy(), elu.numpy(), label="ELU ($\\alpha = 1$)")
axs[0].plot(xs.numpy(), relu.numpy(), label="ReLU")
axs[0].plot(xs.numpy(), selu.numpy(), label="SELU")
axs[0].plot(xs.numpy(), xs.numpy(), label="Linear", linestyle="dashed")

axs[0].set_xlabel("$x$", fontsize=18)
axs[0].set_ylabel("$y$", fontsize=18)

axs[1].set_title("First derivative", fontsize=18)
axs[1].plot(xs.numpy(), sigmoid_grad.numpy(), label="Sigmoid")
axs[1].plot(xs.numpy(), elu_grad.numpy(), label="ELU ($\\alpha = 1$)")
axs[1].plot(xs.numpy(), relu_grad.numpy(), label="ReLU")
axs[1].plot(xs.numpy(), selu_grad.numpy(), label="SELU")
axs[1].plot(xs.numpy(), np.ones_like(xs.numpy()), label="Linear", linestyle="dashed")

axs[1].set_xlabel("$x$", fontsize=18)
axs[1].set_ylabel("$y'$", fontsize=18)

handles, labels = axs[1].get_legend_handles_labels()
#fig.legend(handles, labels, loc='best')
#fig.legend(handles, labels, loc="lower center", fancybox=True, shadow=True, ncol=5, fontsize=16)

# Shrink current axis by 20%
box0 = axs[0].get_position()
axs[0].set_position([box0.x0, box0.y0, box0.width * 0.8, box0.height])

box1 = axs[1].get_position()
axs[1].set_position([box1.x0, box1.y0, box1.width * 0.8, box1.height])

# Put a legend to the right of the current axis
axs[0].legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=11.5)

fig.savefig("activations.pdf", bbox_inches="tight")
```
